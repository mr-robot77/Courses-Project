```python
import matplotlib.pyplot as plt

fig , ax = plt.subplots()
apps = ['chrome','instagram','whatsapp','workout','maps','notes']
hours = [2.2, 3, 1.1, 0.1, 0.85, 0.4]
ax.bar(apps,hours)
ax.set_title("time spent on mobile apps")
ax.set_xlabel("app")
ax.set_ylabel("hour");
```

```python
fig , ax = plt.subplots()

prices = [899,999,1499,1199,1870]

laptops = ["Apple MacBook Air (M1 2020)", "Microsoft Surface Laptop 4","Dell XPS 15 (2020)","HP Spectre x360 (2021)","LG Gram 17 (2021)"]
ax.bar(laptops, prices)
ax.set_title("laptop prices")
ax.set_ylabel("price");
```

```python
fig , ax = plt.subplots()

ax.barh(laptops,prices)
ax.set_title("laptop prices")
ax.set_xlabel("price");
```

```python
fig , ax = plt.subplots(1,2, figsize=(14,5))

ages = ['0-9','10-19','20-29','30-39','40-49','50-59','>60']
population = [222935, 194511, 255757, 315781, 226703, 168848, 174158]

ax[0].bar(ages, population, color='red')
ax[0].set_title("sorted by age")
ax[0].set_ylabel("population")

ages2 = ['30-39', '20-29', '40-49','0-9','10-19','>60','50-59']
population2 = [315781, 255757, 226703, 222935, 194511, 174158, 168848]

ax[1].bar(ages2,population2,color='green')
ax[1].set_title("sorted by population")
ax[1].set_ylabel("population")

fig.suptitle("age distribution in Tabriz");
```

```python
fig , ax = plt.subplots()
ages = ['0-9','10-19','20-29','30-39','40-49','50-59','>60']
female_population = [107770,  96027, 129288, 158169, 111618,  83929,  85231]
male_population = [115165, 98484, 126469, 157612, 115085, 84919, 88927]

ax.bar(ages,male_population, label = 'male')
ax.bar(ages,female_population,bottom=male_population, label='female')
ax.legend()
ax.set_title("age distribution in Tabriz")
ax.set_ylabel("population");
```

```python
import pandas as pd

ages = ['0-9','10-19','20-29','30-39','40-49','50-59','>60']
data = {'male':[115165, 98484, 126469, 157612, 115085, 84919, 88927],
        'female':[107770,  96027, 129288, 158169, 111618,  83929,  85231]}
df = pd.DataFrame(data, index = ages)
df
```

```python
fig , ax = plt.subplots(figsize = (10,4))
df.plot(ax = ax, kind = 'bar')
ax.set_title("age distribution in Tabriz");
```
