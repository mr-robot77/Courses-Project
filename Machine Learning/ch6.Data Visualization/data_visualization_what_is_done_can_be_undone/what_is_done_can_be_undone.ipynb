```python
import matplotlib.pyplot as plt
# TODO: regenerate the plot
days = ["saturday", "sunday", "monday", "tuesday", "wednesday", "thursday", "friday"]
time = [45, 40, 35, 40, 25, 35, 15]
fig , ax = plt.subplots(figsize=(8,6))
# Plot data
ax.plot(days, time, marker='p', markersize=12, markerfacecolor='orange', linestyle='dashed', linewidth=2, color='oldlace')

# Set figure background
fig.set_facecolor('oldlace')

# Set axes title
ax.set_title('minutes i ran last week', fontsize=18, color='mediumvioletred')

# Set axes labels
ax.set_xlabel('day', fontsize=16, color='blueviolet')
ax.set_ylabel('time (minute)', fontsize=16, color='blueviolet')

# Set axes background
ax.set_facecolor('blueviolet')

# Change color of tick labels
plt.xticks(rotation=90)
ax.tick_params(axis='x', colors='mediumvioletred', labelsize=16)
plt.show()
```

    برای ساخته‌شدن فایل <code>result.zip</code> سلول زیر را اجرا کنید. توجه داشته باشید که پیش از اجرای سلول زیر تغییرات اعمال شده در نت‌بوک را ذخیره کرده باشید (<code>ctrl+s</code>) تا در صورت نیاز به پشتیبانی امکان بررسی کد شما وجود داشته باشد.
</font>

```python
import zlib
import zipfile
import joblib

joblib.dump(fig,"fig")
joblib.dump(fig,"fig_t")

def compress(file_names):
    print("File Paths:")
    print(file_names)
    # Select the compression mode ZIP_DEFLATED for compression
    # or zipfile.ZIP_STORED to just store the file
    compression = zipfile.ZIP_DEFLATED
    # create the zip file first parameter path/name, second mode
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            # Add file to the zip file
            # first parameter file to zip, second filename in zip
            zf.write('./' + file_name, file_name, compress_type=compression)


file_names = ["fig", "what_is_done_can_be_undone.ipynb"]
compress(file_names)
```
